```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error,mean_absolute_error
```

```python
student_dataset = pd.read_csv('stud.csv')
student_dataset['total_score'] = student_dataset['math_score']+student_dataset['reading_score']+student_dataset['writing_score']
student_dataset['avg_score'] = student_dataset['total_score']/3
```

```python
student_dataset['performance'] = student_dataset['avg_score'].apply(lambda p:1 if p>75.0 else 0)
student_dataset.head()
```

```python
X=student_dataset.select_dtypes(include='object')
X.head()
```

```python
Y=student_dataset['performance']
Y.head()
```

```python
X=X.replace({
    'gender':{'female':0,'male':1},
    'lunch':{'standard':1,'free/reduced':0},
    'test_preparation_course':{'none':0,'completed':1},
    'race_ethnicity':{'group A':0,'group B':1,'group C':2,'group D':3,'group E':4},
    'parental_level_of_education':{"master's degree":0,"bachelor's degree":1,"some college":2,"associate's degree":3,"high school":4,"some high school":5}
})
X.head()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=50)
```

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train,Y_train)
```

```python
X_train_pred = classifier.predict(X_train)
X_test_pred = classifier.predict(X_test)
train_acc = accuracy_score(X_train_pred,Y_train)
test_acc = accuracy_score(X_test_pred,Y_test)

print('Training accuracy: ',train_acc)
print('Testing accuracy: ',test_acc)
```
