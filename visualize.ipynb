```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
student_dataset = pd.read_csv('stud.csv')
student_dataset.head(10)
```

```python
student_dataset['total_score'] = student_dataset['math_score']+student_dataset['reading_score']+student_dataset['writing_score']
student_dataset['avg_score'] = student_dataset['total_score']/3
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data=student_dataset,x='test_preparation_course',y='avg_score')
```

```python
sns.barplot(data=student_dataset,x='lunch',y='avg_score')
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=student_dataset,x='parental_level_of_education',y='avg_score')
```

```python
sns.barplot(data=student_dataset,x='race_ethnicity',y='avg_score')
```

```python
fig,axs = plt.subplots(1,2,figsize=(20,10))
plt.subplot(121)
sns.histplot(data=student_dataset,x='avg_score')
plt.subplot(122)
sns.histplot(data=student_dataset,hue='gender',x='avg_score',kde=True)
```

```python
fig,axs = plt.subplots(1,3,figsize=(20,10))
plt.subplot(131)
plt.title('Female')
sns.histplot(data=student_dataset[student_dataset['gender']=='female'],x='avg_score',hue='test_preparation_course',kde=True)
plt.subplot(132)
plt.title('Male')
sns.histplot(data=student_dataset[student_dataset['gender']=='male'],x='avg_score',hue='test_preparation_course',kde=True)
plt.subplot(133)
sns.histplot(data=student_dataset,hue='test_preparation_course',x='avg_score',kde=True)
#test_preparation_course
```

```python
fig,axs = plt.subplots(1,3,figsize=(20,10))
plt.subplot(131).set_title('lunch')
sns.histplot(data=student_dataset,x='avg_score',hue='lunch',kde=True)
plt.subplot(132).set_title('Female')
sns.histplot(data=student_dataset[student_dataset['gender']=='female'],x='avg_score',hue='lunch',kde=True)
plt.subplot(133).set_title('Male')
sns.histplot(data=student_dataset[student_dataset['gender']=='male'],x='avg_score',hue='lunch',kde=True)
```
